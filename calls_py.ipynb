```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

df_calls = pd.read_csv('calls.csv', sep = ',')
display(df_calls.head(500))
df_calls.info()
```

```python
df_calls.describe()
```

```python
df = df_calls.head(3000)
plt.figure(figsize=(18,8))
plt.plot('time','val', data = df) # важная строка, собственно, строит график
plt.grid() # сетка
# plt.legend(loc='upper left') 
plt.title('Электрокардиограмма')
plt.xlabel('Время')
plt.ylabel('Сердце')
plt.show()

```

```python
R_prong = [] # зубцы
time_R = []  # время зубцов
t=0
while t < df['time'].size:
    if df.loc[t,'val'] > 125:
        max = 0
        t_max = t
        while df.loc[t,'val'] > 120:
            if df.loc[t,'val'] > max:
                max = df.loc[t,'val']
                t_max = df.loc[t,'time']
            t = t+1
        R_prong.append(max)
        time_R.append(t_max)

    t = t+1
display(R_prong)
time_R
            
    
```

```python
plt.figure(figsize=(18,8))
plt.plot('time','val', data = df) # важная строка, собственно, строит график
plt.grid() # сетка
# plt.legend(loc='upper left') 
plt.title('Электрокардиограмма')
plt.xlabel('Время')
plt.ylabel('Сердце')

plt.scatter(time_R, R_prong, color = 'r') #строим точки пиков
plt.show()
```
